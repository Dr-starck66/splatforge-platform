## 2. Préparation des données
Téléchargez votre vidéo ou vos images dans le dossier `input_data` à gauche dans l'onglet fichiers.

```python
from pathlib import Path
input_dir = Path("input_data")
output_dir = Path("data/outputs")
input_dir.mkdir(exist_ok=True)
output_dir.mkdir(parents=True, exist_ok=True)

print(f"✅ Dossiers prêts. Déposez vos fichiers dans : {input_dir.absolute()}")
```

## 3. Lancement du Pipeline
Exécutez cette cellule pour lancer la reconstruction.

```python
import sys
sys.path.append(os.getcwd())
from app.pipeline import process_dataset

SESSION_ID = "colab_test"

try:
    result = process_dataset(
        input_folder=str(input_dir),
        output_folder=str(output_dir),
        session_id=SESSION_ID,
        enable_3dgs=True
    )
    print("\n🎉 Reconstruction terminée avec succès !")
    print(f"Résultats disponibles dans : {output_dir / SESSION_ID}")
except Exception as e:
    print(f"\n❌ Erreur lors du traitement : {e}")
```

## 4. Télécharger les résultats
Compresse et télécharge le fichier `.splat` final.

```python
from google.colab import files
import shutil

output_zip = f"{SESSION_ID}_results.zip"
shutil.make_archive(SESSION_ID, 'zip', output_dir / SESSION_ID)
files.download(f"{SESSION_ID}.zip")
```
